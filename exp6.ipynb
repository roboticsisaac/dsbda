```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Post Month', y='Page total likes', data=df)
plt.xlabel('Post Month')
plt.ylabel('Page Total Likes')
plt.title('Page Total Likes by Post Month')

plt.show()
```
